```python
import re
pat = re.compile('(info|atc)_(code|type|repeated)\d')
dropcols = [col for col in asrs if pat.match(col)]
```

```python
all_pds = []
for info_type in ['atc', 'info']:
    for i in range(4):
        print(f"{info_type}_{i}")
        curr = asrs.loc[asrs[f"{info_type}_type{i}"] == 'TRACON', :].copy()
        tracon_code = curr[f"{info_type}_code{i}"].copy()
        curr.drop(dropcols, axis = 1, inplace = True)
        curr['tracon_code'] = tracon_code
        all_pds.append(curr)
```

```python
def convert_to_words(row, col = 'narrative'):
    s = row[col]
    if isinstance(s, float) and np.isnan(s):
        s = ''
    res = np.array(re.split('[( | ;|\. |\.$]', s))
    res = res[res != '']
    return res
```

```python
asrs = asrs.iloc[1:]
```

```python
asrs['narrative'][~(asrs['narrative'].str.isupper())]
```

```python
narrative_split = asrs.apply(lambda x: convert_to_words(x, 'narrative'), axis = 1)
```

```python
from tqdm import tqdm_notebook
```

```python
dbs = ['casa', 'faa', 'iata_iaco']
for list_of_words in tqdm_notebook(narrative_split):
    for word in list_of_words:
        for db in dbs:
            if word in dbs_dict[db]:
                dbs_dict[db][word] += 1
```

```python
casa_cts = pd.DataFrame(dbs_dict['casa'], index = [0]).T
casa_cts.rename({0: 'counts'}, inplace = True, axis = 1)
casa_cts.sort_values(by = 'counts', inplace = True)
casa_cts.to_csv('results/casa_cts')
```

```python
faa_cts = pd.DataFrame(dbs_dict['faa'], index = [0]).T
faa_cts.rename({0: 'counts'}, inplace = True, axis = 1)
faa_cts.sort_values(by = 'counts', inplace = True)
faa_cts.to_csv('results/faa_cts')
```
