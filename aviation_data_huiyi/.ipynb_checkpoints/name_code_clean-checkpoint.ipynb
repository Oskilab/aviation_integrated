```python
import pandas as pd
import numpy as np
import ftfy
```

```python
ntsb_code = pd.read_csv("NTSB_airportcode.csv")
ntsb_name = pd.read_csv("NTSB_airportname.csv")
airport_code = pd.read_csv("airports.csv")
```

```python
def clean(col): 
    toReturn = []
    for string in col:
        toReturn.append(''.join([c.upper() for c in string if c.isalpha() or c.isnumeric()]))
    return toReturn
```

```python
# Fix unicode

code_org = [str(d) for d in ntsb_code["airportcode"]]
code_fix_unicode = []
code_failed = []
for code in code_org:
    try:
        code_fix_unicode.append(ftfy.fix_text(code))
    except:
        code_failed.append(str(code))  
len(code_failed) == 0
```

```python
# Fix unicode

name_org = [str(d) for d in ntsb_name["airportname"]]
name_fix_unicode = []
name_failed = []
for code in name_org:
    try:
        name_fix_unicode.append(ftfy.fix_text(code))
    except:
        name_failed.append(str(code))  
len(name_failed) == 0
```

```python
ntsb_code["airportcode"] = clean(code_fix_unicode)
ntsb_name["airportname"] = [string.strip() for string in name_fix_unicode]
```

```python
airport_code['ident'] = [string.upper().strip() for string in airport_code['ident']]
```

### Build a dic for code

```python
# round 1

code_reference = [string.upper().strip() for string in set(airport_code['ident'])]

code_matched = []
code_not_matched = []
for airportcode in set(ntsb_code["airportcode"]): 
    if airportcode in code_reference: 
        code_matched.append(airportcode)
    else: 
        code_not_matched.append(airportcode)
```

```python
len(code_matched), len(code_not_matched)
```

```python
code_not_matched
```

```python
# round 2 - some code in reference have a one-letter prefix

def clean_first_letter(col): 
    toReturn = []
    for string in col: 
        try: 
            toReturn.append(string[1:])
        except: 
            pass
    return toReturn

code_reference2 = clean_first_letter(code_reference)

code_matched2 = []
code_not_matched2 = []
for airportcode in code_not_matched: 
    if airportcode in code_reference2: 
        code_matched2.append(airportcode)
    else: 
        code_not_matched2.append(airportcode)
```

```python
len(code_matched2), len(code_not_matched2)
```

```python
code_not_matched2
```

```python
# round 3 - risky, see if some code in data has its last letter missing

def clean_last_letter(col): 
    toReturn = []
    for string in col: 
        try: 
            toReturn.append(string[:-1])
        except: 
            pass
    return toReturn

code_reference3 = clean_last_letter(code_reference)

code_matched3 = []
code_not_matched3 = []
for airportcode in code_not_matched2: 
    if airportcode in code_reference3: 
        code_matched3.append(airportcode)
    else: 
        code_not_matched3.append(airportcode)
```

```python
len(code_matched3), len(code_not_matched3)
```

```python
code_matched
```

```python
# Build dictionary (code: index in airport_code)
# trusted: code_matched, code_matched2
# risky: code_matched3

code_dict_trusted = dict()
for code in code_matched: 
    ind = airport_code.index[airport_code['ident'] == code].tolist()[0]
    code_dict_trusted.update({code: ind})
```

```python
clean_first = [string[1:] for string in airport_code['ident']]

for code in code_matched2: 
    clean_first_match = [item == code for item in clean_first]
    ind_lst = airport_code.index[clean_first_match].tolist()
    try: 
        ind = ind_lst[0]
        code_dict_trusted.update({code: ind})
    except:
        print(ind_lst)
```

```python
print("percent of code matched with airport data, trusted: " + str(len(code_dict_trusted) / len(ntsb_code)))
```

```python
code_dict_risky = dict()

clean_last = [string[:-1] for string in airport_code['ident']]

for code in code_matched3: 
    clean_last_match = [item == code for item in clean_last]
    ind_lst = airport_code.index[clean_last_match].tolist()
    try: 
        ind = ind_lst[0]
        code_dict_risky.update({code: ind})
    except:
        print(ind_lst)
```

```python
print("percent of code matched with airport data, risky: " + str(len(code_dict_risky) / len(ntsb_code)))
```

### Build a dic for name

```python
def clean_paren(col):
    toReturn = []
    for string in col: 
        string = string.upper().strip()
        left = string.find('(')
        right = string.find(')')
        if left != -1 and right != -1: 
            try: 
                new_string = string[:left] + string[right+1:]
            except:
                new_string = string[:left]
            toReturn.append(new_string)
        elif left != -1: 
            new_string = string[:left]
            toReturn.append(new_string)
        else: 
            toReturn.append(string)
    return toReturn
```

```python
ntsb_name["airportname_cleanedparen"] = clean_paren(name_fix_unicode)
ntsb_name["airportname_cleanedletter"] = clean(clean_paren(name_fix_unicode))
```
